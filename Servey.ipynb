```python
Java = df[df['LanguageHaveWorkedWith'] == 'Java']
Java_count = len(Java)
total_rows = len(df)
percentage_java = (Java_count / total_rows) * 100
print(percentage_java)
```

```python
Go = df[df['LanguageHaveWorkedWith'] == 'Go']
Go_count = len(Go)
total_rows = len(df)
percentage_go = (Go_count / total_rows) * 100
print(percentage_go)
```

```python
Cpp = df[df['LanguageHaveWorkedWith'] == 'C++']
Cpp_count = len(Cpp)
total_rows = len(df)
percentage_cpp = (Cpp_count / total_rows) * 100
print(percentage_cpp)
```
